Yahoo Finance is a good option, but it is only possible to retrieve data daily, monthly, yearly ...

```python
import yfinance as yf
```

```python
BTC_Ticker = yf.Ticker("BTC-USD")
BTC_Data = BTC_Ticker.history(period="max")
BTC_Data
```

```python
# @title Open

from matplotlib import pyplot as plt
BTC_Data['Open'].plot(kind='line', figsize=(8, 4), title='Open')
plt.gca().spines[['top', 'right']].set_visible(False)
```

## Coin Gecko API

This API is also a great one, but the option to fetch data more frequently than daily is difficult to approach.

```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
cg = CoinGeckoAPI()
parameters = {
         'vs_currency': 'usd',
         'order': 'market_cap_desc',
         'per_page': 100,
         'page': 1,
         'sparkline': False,
         'locale': 'en'
         }
coin_market_data = cg.get_coins_markets(**parameters)
```

```python
df = pd.DataFrame(coin_market_data)
df = df.drop(['id', 'symbol', 'image', 'circulating_supply', 'ath', 'atl',
        'market_cap_change_24h','market_cap_change_percentage_24h', 'fully_diluted_valuation', 'ath_date', 'ath_change_percentage',
        'atl_change_percentage', 'atl_date', 'roi'],  axis = 1)
df
```
